```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("./MyPhoto.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image.shape
```

```python
filter = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
```

```python
r, g, b = cv2.split(image)
```

```python
r_sharpened = cv2.filter2D(r, -1, filter)
g_sharpened = cv2.filter2D(g, -1, filter)
b_sharpened = cv2.filter2D(b, -1, filter)
```

```python
sharpened_image = cv2.merge([r_sharpened, g_sharpened, b_sharpened])
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_image)
plt.title('Sharpened')
```

```python
image_blurred_gaussian_3 = cv2.GaussianBlur(image, (3, 3), 0)
image_blurred_gaussian_5 = cv2.GaussianBlur(image, (5, 5), 0)
image_blurred_gaussian_7 = cv2.GaussianBlur(image, (7, 7), 0)
image_blurred_gaussian_9 = cv2.GaussianBlur(image, (9, 9), 0)
image_blurred_gaussian_11 = cv2.GaussianBlur(image, (11, 11), 0)
image_blurred_gaussian_13 = cv2.GaussianBlur(image, (13, 13), 0)
image_blurred_gaussian_15 = cv2.GaussianBlur(image, (15, 15), 0)
```

```python
r_blurred_gaussian_5_sharpened, g_blurred_gaussian_5_sharpened, b_blurred_gaussian_5_sharpened = cv2.split(image_blurred_gaussian_5)
r_blurred_gaussian_5_sharpened = cv2.filter2D(r_blurred_gaussian_5_sharpened, -1, filter)
g_blurred_gaussian_5_sharpened = cv2.filter2D(g_blurred_gaussian_5_sharpened, -1, filter)
b_blurred_gaussian_5_sharpened = cv2.filter2D(b_blurred_gaussian_5_sharpened, -1, filter)

image_blurred_gaussian_5_sharpened = cv2.merge([r_blurred_gaussian_5_sharpened, g_blurred_gaussian_5_sharpened, b_blurred_gaussian_5_sharpened])
```

```python
plt.figure(figsize=(30, 15))

plt.subplot(2, 4, 1)
plt.imshow(image)
plt.title('Original')

plt.subplot(2, 4, 2)
plt.imshow(image_blurred_gaussian_5_sharpened)
plt.title('Gaussian Blurred (5,5) and Sharpened')

plt.subplot(2, 4, 3)
plt.imshow(image_blurred_gaussian_3)
plt.title('Gaussian Blurred (3,3)')

plt.subplot(2, 4, 4)
plt.imshow(image_blurred_gaussian_5)
plt.title('Gaussian Blurred (5,5)')

plt.subplot(2, 4, 5)
plt.imshow(image_blurred_gaussian_9)
plt.title('Gaussian Blurred (9,9)')

plt.subplot(2, 4, 6)
plt.imshow(image_blurred_gaussian_11)
plt.title('Gaussian Blurred (11,11)')

plt.subplot(2, 4, 7)
plt.imshow(image_blurred_gaussian_13)
plt.title('Gaussian Blurred (13,13)')

plt.subplot(2, 4, 8)
plt.imshow(image_blurred_gaussian_15)
plt.title('Gaussian Blurred (15,15)')
```
